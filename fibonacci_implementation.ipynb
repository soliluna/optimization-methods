```python
import math as m
```

```python
# объявление функций
def my_func(x):
    """Возвращает значение функции в заданной точке."""
    a_0 = 5
    a_1 = 3
    a_2 = 7
    a_3 = 1
    return x ** 4 + a_3 * x ** 3 + a_2 * x ** 2 + a_1 * x + a_0

def binet_formula(n):
    """Возвращает значение n-го члена ряда Фибоначчи."""
    return (m.pow((1 + m.sqrt(5)) / 2, n) - m.pow((1 - m.sqrt(5) / 2), n)) / m.sqrt(5)
```

## Метод Фибоначчи

```python
# начальный этап алгоритма
a, b = -3, 2
epsilon = 0.0001
n = 0

while True:
    n += 1
    F_n = binet_formula(n)
    if F_n > ((b - a) / epsilon):
        break

x_1 = a + binet_formula(n - 2) / F_n * (b - a)
x_2 = a + binet_formula(n - 1) / F_n * (b - a)
f_1 = my_func(x_1)
f_2 = my_func(x_2)
k = 1

print("k = %-3.d a = %-10.5f b = %-10.5f x1 = %-10.5f x2 = %-10.5f "
  "f(x1) = %-10.5f f(x2) = %-10.5f"
  % (k, a, b, x_1, x_2, f_1, f_2)
  )

if f_1 > f_2:
    a = x_1
else:
    b = x_2

# основной этап алгоритма
while not k == n - 1: 
    k += 1

    if f_1 > f_2:
        x_1 = x_2
        x_2 = a + (binet_formula(n - k - 1) / binet_formula(n - k)) * (b - a)
        f_1, f_2 = my_func(x_1), my_func(x_2)
    else:
        x_2 = x_1
        x_1 = a + binet_formula(n - k - 2) / binet_formula(n - k) * (b - a)
        f_1, f_2 = my_func(x_1), my_func(x_2)

    print("k = %-3.d a = %-10.5f b = %-10.5f x1 = %-10.5f x2 = %-10.5f "
        "f(x1) = %-10.5f f(x2) = %-10.5f"
        % (k, a, b, x_1, x_2, f_1, f_2)
      )

    if f_1 > f_2:
        a = x_1
    else:
        b = x_2  
```
