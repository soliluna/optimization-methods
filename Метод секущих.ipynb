## Метод секущих

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# объявление функции
def my_func(x):
    """Возвращает значение функции в заданной точке."""
    a_0 = 5 
    a_1 = 3
    a_2 = 7
    a_3 = 1
    return x ** 4 + a_3 * x ** 3 + a_2 * x ** 2 + a_1 * x + a_0


# объявление производной функции
def my_func_derivative(x):
    """Возвращает значение производной функции в заданной точке."""
    a_0 = 5 
    a_1 = 3
    a_2 = 7
    a_3 = 1
    return 4* x ** 3 + 3 * a_3 * x ** 2 + 2* a_2 * x + a_1
```

```python
# реализация алгоритма для 6 случаев точности
for n in range(1, 7):
    a, b = -3, 2
    epsilon = 10 ** (-n)
    k = 0

    while True:
        k += 1
        f_a = my_func_derivative(a)
        f_b = my_func_derivative(b)
        z = b - (f_b * (b - a)) / (f_b - f_a)
        f_z = my_func_derivative(z)

        print(
            "k = %-3d a = %-10.5f b = %-10.5f f'(a) = %-10.5f f'(b) = %-10.5f "
            "z = %-10.5f f'(z) = %-10.5f"
            % (k, a, b, f_a, f_b, z, f_z))

        if f_z > epsilon:
            if f_z < 0:
                a = z
            else:
                b = z
        else:
            break

    # вывод результата оценки минимума
    print('\nКоличество итераций: %d' % k)
    print('Оценка х*: %.3f' % z)
    print('Оценка f(x*): %.3f' % (my_func(z)))
    print('-' * 106)
```

```python
# оптимизация алгоритма для 6 случаев точности
for n in range(1, 7):
    a, b = -3, 2
    epsilon = 10 ** (-n)
    k = 0

    while True:
        k += 1
        f_a = my_func_derivative(a)
        f_b = my_func_derivative(b)
        z = b - (f_b * (b - a)) / (f_b - f_a)
        f_z = my_func_derivative(z)

        print(
            "k = %-3d a = %-10.5f b = %-10.5f f'(a) = %-10.5f f'(b) = %-10.5f "
            "z = %-10.5f f'(z) = %-10.5f"
            % (k, a, b, f_a, f_b, z, f_z))

        if f_z > epsilon:
            b = z
        else:
            break

    # вывод результата оценки минимума
    print('\nКоличество итераций: %d' % k)
    print('Оценка х*: %.3f' % z)
    print('Оценка f(x*): %.3f' % (my_func(z)))
    print('-' * 106)
```

## График производной

```python
# график производной
fig, ax = plt.subplots(figsize=(10, 7))
ax.axis([-2.5, 2.5, -50.5, 50.5])
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

fig.suptitle('Алгоритм поиска минимума методом секущих',
             fontsize='x-large', fontweight='semibold'
            )
ax.set_xlabel('x', fontsize=10, fontweight=5)
ax.set_ylabel("f'(x)", fontsize=10, fontweight=5, rotation=1)

x = np.arange(-2, 2, 0.1)
y = [my_func_derivative(i) for i in x]

plt.plot(x, y, color='mediumvioletred')

plt.show()
```
